# Rough analysis of the data for the credit card fraud detection poblem

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../creditcard.csv')
```

```python
data.describe()
```

As it can be seen in the *Class* column, only 0.1727% of the data is fraud.
PCA variables are centered in 0 and have an order of magnitude of O(10).

```python
data['Class'].value_counts()
```

```python
clean = data[data['Class'] == 0]
fraud = data[data['Class'] == 1]
```

## Amount analysis

```python
fig, axs = plt.subplots(1,2, figsize=(20, 5))

clean['Amount'].plot.hist(bins=100, logy=True, ax = axs[0], label='Clean' )
fraud['Amount'].plot.hist(bins=10, logy=True, ax = axs[0], label='Fraud' )
axs[0].set(xlabel='Amount')
axs[0].legend(loc='best')

fraud['Amount'].sort_values().reset_index(drop=True).plot(ax=axs[1], c='tab:orange')
axs[1].set(xlabel='Fraud transactions sorted by amount (ascending)', ylabel='Amount')

#data2 = data['Amount'].sort_values().reset_index(drop=True).cumsum()
#data2.apply(lambda x: x / data2.iloc[-1] * 100).plot(ax = axs[1], grid = True)
```

## Time analysis

```python
fig, axs = plt.subplots(1,2, figsize=(20, 5))

clean['Time'].apply(lambda x: x / 3600.0).hist(bins = 20, ax = axs[0])
fraud['Time'].apply(lambda x: x / 3600.0).hist(bins = 20, ax = axs[1], color='tab:orange')
```

```python
fig, ax = plt.subplots()
ax.plot(data['Time'].apply(lambda x: x / 3600.0), data['Class'].apply(lambda x: 1 - x).cumsum())
ax2 = ax.twinx()
ax2.plot(data['Time'].apply(lambda x: x / 3600.0), data['Class'].cumsum(), color='tab:orange')
ax3 = ax.twinx()
ax3.plot(data['Time'].apply(lambda x: x / 3600.0), data['Time'], color='g')
```

## V variables analysis

```python
for i in np.arange(1,29,1):
    fig, axs = plt.subplots(figsize=(20, 5))
    clean['V'+str(i)].hist(ax=axs, bins=30)
    ax2 = axs.twinx()
    fraud['V'+str(i)].hist(ax=ax2, color='tab:orange', alpha=0.5)
```
