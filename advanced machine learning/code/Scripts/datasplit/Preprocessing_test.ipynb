```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from preprocessing import get_processed_data
```

```python
df = get_processed_data("../../Dataset/creditcard.csv")
```

```python
df.describe()
```

```python
clean = df[df['Class'] == 0]
fraud = df[df['Class'] == 1]

clean = clean.reset_index(drop=True)
fraud = fraud.reset_index(drop=True)
```

```python
fig,ax = plt.subplots()

clean['LogAmount'].plot.hist(bins=100, logy=True, ax=ax)
fraud['LogAmount'].plot.hist(bins=20, logy=True, ax=ax)

ax.set(xlabel='Normalized log withdrawn amount')
```

```python
fig, axs = plt.subplots(1, 3, figsize=(20, 5))
#df['TimeComponent1'].plot(c='g', ax=axs[0])
#df['TimeComponent2'].plot(c='g', ax=axs[0])
axs[0].plot(np.linspace(0,4*np.pi), np.sin(np.linspace(0,4*np.pi)), c='g')
axs[0].plot(np.linspace(0,4*np.pi), np.cos(np.linspace(0,4*np.pi)), c='g')

clean['TimeComponent1'].plot(c='b', ax=axs[1])
clean['TimeComponent2'].plot(c='b', ax=axs[1])

fraud['TimeComponent1'].plot(c='orange')
fraud['TimeComponent2'].plot(c='orange')
```
